Research Question: How does exposure to extreme heat events affect hospital admissions for cardiovascular disease in the United States, while controlling for air quality and access to green spaces?
Hypothesis: Exposure to extreme heat events, as measured by high temperature and heat index values, will increase the risk of hospital admissions for cardiovascular disease, with potentially greater effects in certain regions and for certain demographic groups. This relationship will be influenced by air pollution and access to green spaces, with higher levels of air pollution and lack of access to green spaces exacerbating the health effects of extreme heat events.

The graph above shows that some states of the US have experienced more warming than others. Going by the average daily max temperature Florida, Louisiana and Texas were by far the warmest states while the coldest states were Wyoming, Vermont and Maine.

```python
df['median_income'] = df['median_income'].str.replace(',', '').astype(float)
```

```python
sns.scatterplot(x='median_income', y='Hospitalizations', data=df)
```

```python
sns.regplot(x='median_income', y='Hospitalizations', data=df)
```


The graph above shows the income distribution of individuals who were admitted to the hospital for cardiovascular disease. It shows that individuals with higher income are less likely to be admitted. This finding is consistent with previous research that has shown a link between socioeconomic status and cardiovascular disease. Individuals with lower socioeconomic status, including lower income levels, may face greater barriers to accessing healthcare, have higher levels of stress, and engage in more unhealthy behaviors such as smoking and poor diet, which can increase their risk of cardiovascular disease.

```python
sns.scatterplot(x='Air_quality', y='Hospitalizations', data=df)
```

```python
sns.regplot(x='Air_quality', y='Hospitalizations', data=df)
```

There appears to be a moderate relationship between air quality and the number of hospital admissions for cardiovascular diseases. 
One possible explanation for this trend is that poor air quality is a risk factor for cardiovascular diseases, which can lead to an increased number of hospital admissions. Air pollution, particularly from fine particulate matter and nitrogen dioxide, has been linked to an increased risk of heart disease and stroke. Long-term exposure to air pollution can cause inflammation and damage to blood vessels, which can contribute to the development of cardiovascular diseases.

Another possibility is that people living in areas with poor air quality may also be exposed to other environmental and lifestyle risk factors for cardiovascular diseases, such as poor diet, lack of physical activity, and exposure to secondhand smoke. These factors could contribute to the increased number of hospital admissions for cardiovascular diseases.

It is also possible that the relationship between air quality and hospital admissions for cardiovascular diseases is influenced by other factors, such as the age and health status of the population, access to healthcare, and socioeconomic status.
